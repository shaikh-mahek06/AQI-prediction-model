```python
def replace_outliers(df):
    for column in df.select_dtypes(include=['number']).columns:
        Q1 = df[column].quantile(0.25)  # Calculate 25th percentile
        Q3 = df[column].quantile(0.75)  # Calculate 75th percentile
        IQR = Q3 - Q1                   
        lb = Q1 - 1.5 * IQR             
        ub = Q3 + 1.5 * IQR            
        df[column] = df[column].apply(
            lambda x: Q1 if x < lb else (Q3 if x > ub else x)
        )  # Replace outliers
    return df
```

```python
df=replace_outliers(df)
```
